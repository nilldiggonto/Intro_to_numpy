```python
np.random.seed(123)
m = np.random.normal(10,2,20).reshape(2,10)
```

```python
m
```

```python
a = m[0]
```

```python
a
```

```python
b =m[1]
```

```python
b
```

```python
###
```

```python
plt.figure(figsize=(12,8))
plt.scatter(a,b)
plt.show()
```

```python
####### Linear Regression
```

```python
reg1 =  np.polyfit(a,b,1)
```

```python
reg1
```

```python
x  = np.linspace(min(a),max(a),num=100)
y = np.polyval(reg1,x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(12,8))
plt.scatter(a,b,label='Data')
plt.plot(x,y,'--',label='linear')
plt.legend()
plt.show()
```

```python
#quadratic regression
```

```python
reg2 = np.polyfit(a,b,2)
reg2
```

```python
reg3 = np.polyfit(a,b,3)
reg3
```

```python
plt.figure(figsize=(12,8))
plt.scatter(a,b,label='Data')
plt.plot(x,y,'--',label='linear')
plt.plot(x,np.polyval(reg2,x),'--',label='Quadratic')
plt.plot(x,np.polyval(reg3,x),'.',label='Cubic')
plt.legend()
plt.show()
```
